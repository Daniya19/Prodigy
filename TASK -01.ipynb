```python
columns=['LotFrontage','GarageYrBlt','MasVnrArea']
for column in columns:
    df[column]=df[column].fillna(df[column].median())
```

```python
categorical_columns=['Alley','MasVnrType','BsmtQual','BsmtExposure','BsmtFinType1','BsmtFinType2','Electrical','FireplaceQu','GarageType',
                     'GarageYrBlt'	,'GarageFinish','GarageQual','GarageCond','PoolQC','Fence','MiscFeature']
for column in categorical_columns:
    df[column]=df[column].fillna(df[column].mode()[0])

```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
features=['GrLivArea','FullBath','BedroomAbvGr']
x=df[features]
y=df['SalePrice']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

```python
import seaborn as sns
for i in df.select_dtypes(include="number").columns:
    sns.histplot(data=df,x=i)
    plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
print("MSE= ",mse)
r2=r2_score(y_test,y_pred)
print("R Squared= ",r2)
```
